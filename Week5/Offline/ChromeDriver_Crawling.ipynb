```python
import pyperclip
from selenium.webdriver.common.keys import Keys
```

```python
url = "https://www.naver.com/"
wd.get(url)

naver_id = "qkrqldtjqj"
naver_pw = "navereptmtjq1"

login_button = wd.find_element_by_css_selector("#account > a > i")
login_button.click()

input_id = wd.find_element_by_css_selector("#id")
input_id.click()
pyperclip.copy(naver_id)
input_id.send_keys(Keys.CONTROL, 'v')
time.sleep(1)

input_pw = wd.find_element_by_css_selector("#pw")
input_pw.click()
pyperclip.copy(naver_pw)
input_pw.send_keys(Keys.CONTROL, 'v')
time.sleep(1) 
```
