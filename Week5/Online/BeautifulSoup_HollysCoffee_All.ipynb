```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
def hollysCrawling(pageNum):
  store_name = []
  store_address = []
  store_tel = []

  for n in range(0, pageNum):
    hollysURL = 'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=%d&sido=&gugun=&store=' % (n + 1)
    hollysHTML = urllib.request.urlopen(hollysURL).read()

    hollysSoap = BeautifulSoup(hollysHTML, 'html.parser')
    hollysTBody = hollysSoap.find('tbody')
    hollysTR = hollysTBody.find_all('tr')

    for tr in hollysTR:
      hollysTD = tr.find_all('td')
      store_name.append(hollysTD[1].string)
      store_address.append(hollysTD[3].string)
      store_tel.append(hollysTD[5].string)

  Hollys_dic = {'매장이름':store_name, '주소':store_address, '전화번호':store_tel}
  Hollys_store = pd.DataFrame(Hollys_dic)

  return Hollys_store
```

```python
Hollys_store = hollysCrawling(54)
Hollys_store
```

```python
Hollys_store.to_excel('hollys_coffee.xlsx', sheet_name='shop')
```
