```python
import pandas as pd
import folium
```

```python
filePath = '/content/drive/MyDrive/SSU_데이터 사이언스/CoffeeBean.xlsx'
cb = pd.read_excel(filePath)
cb
```

```python
cb.drop('Unnamed: 0', axis = 1, inplace = True)
cb
```

```python
cb['address']
```

```python
addr = []

for address in cb['address']:
  addr.append(address.split())

addr
```

```python
for add in addr:
  if add[0] == '서울':
    add[0] = '서울특별시'
  elif add[0] == '서울시':
    add[0] = '서울특별시'
  elif add[0] == '경기':
    add[0] = '경기도'
  elif add[0] == '경남':
    add[0] = '경상남도'
  elif add[0] == '광주':
    add[0] = '광주광역시'
  elif add[0] == '대전':
    add[0] = '대전광역시'
  elif add[0] == '부산':
    add[0] = '부산광역시'
  elif add[0] == '부산시':
    add[0] = '부산광역시'
  elif add[0] == '대전':
    add[0] = '대전광역시'
  elif add[0] == '대전시':
    add[0] = '대전광역시'
  elif add[0] == '충북':
    add[0] = '충청북도'
  elif add[0] == '충남':
    add[0] = '충청남도'
  elif add[0] == '인천':
    add[0] = '인천광역시'
  elif add[0] == '인천시':
    add[0] = '인천광역시'
  elif add[0] == '제주시':
    add[0] = '제주특별자치도'
  elif add[0] == '제주':
    add[0] = '제주특별자치도'

addr
```

```python
unique = []

for add in addr:
  if add[0] not in unique:
      unique.append(add[0])

unique
```

```python
# 혹은 set를 쓰면 중복을 허용하지 않는 dataset 완성
set(unique)
```
