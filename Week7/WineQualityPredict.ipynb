```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
redWinePath = 'winequality-red.csv'

redDataFrame = pd.read_csv(redWinePath, sep=';')
redDataFrame
```

```python
whiteWinePath = 'winequality-white.csv'

whiteDataFrame = pd.read_csv(whiteWinePath, sep=';')
whiteDataFrame
```

```python
redDataFrame.columns == whiteDataFrame.columns
```

```python
redDataFrame.insert(0, column = 'type', value = 'red')
redDataFrame
```

```python
whiteDataFrame.insert(0, column = 'type', value = 'white')
whiteDataFrame
```

```python
totalWineDF = pd.concat([redDataFrame, whiteDataFrame])
totalWineDF
```

```python
print(totalWineDF.shape)
print(totalWineDF.info())
```

```python
totalWineDF.columns = totalWineDF.columns.str.replace(' ', '_')
totalWineDF
```

```python
totalWineDF.isnull().sum()
```

```python
totalWineDF.describe()
```

```python
totalWineDF.groupby('type')['quality'].describe()
```

```python
import matplotlib.pyplot as plt

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 12

data1 = totalWineDF[totalWineDF['type'] == 'red'].quality
data2 = totalWineDF[totalWineDF['type'] == 'white'].quality

flg, ax = plt.subplots()

ax.boxplot([data1, data2], notch = False, whis = 1.5)
ax.set_ylim(-1.0, 10.0)
ax.set_xlabel('Wine Type')
ax.set_xlabel('Quality')
```

```python
from scipy import stats
```

```python
reqWineQuality = totalWineDF.loc[totalWineDF['type'] == 'red', 'quality']
whiteWineQuality = totalWineDF.loc[totalWineDF['type'] == 'white', 'quality']

stats.ttest_ind(reqWineQuality, whiteWineQuality, equal_var=False)
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
results = smf.ols('quality ~ fixed_acidity + volatile_acidity + citric_acid + residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide + density + pH + sulphates + alcohol', data=totalWineDF).fit()
results.summary()
```
